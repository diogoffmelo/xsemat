# Resolução EDO's

Vamos resolver o PVI (problema de valor inicial) da circunfrência


```python
def dy(t, y):
    return y[1], -y[0]


t = numpy.linspace(0, 10, 100)

sol = integrate.solve_ivp(dy, [0, 10], [0, 1], t_eval=t)

pyplot.plot(sol.t, sol.y[0, :], label='x(t)')
pyplot.plot(sol.t, sol.y[1, :], label='y(t)')

pyplot.legend()
```

```python
pyplot.plot(sol.y[1, :], sol.y[0, :], label='x(t)')
```

# Problema da queda livre com atrito

Baseado em <https://ipython-books.github.io/123-simulating-an-ordinary-differential-equation-with-scipy/>

```python
M = 1  # Massa
K = 1  # Arrasto
G = 9.81  # Gravidade

y0 = [0, 0, 5, 5]

def dy(t, y):
    x1, x2, v1, v2 = y
    return v1, v2, -K / M * v1, -K / M * v2 - G


t = numpy.linspace(0, 1, 100)
sol = integrate.solve_ivp(dy, [0, 10], y0, t_eval=t)

pyplot.plot(sol.y[0, :], sol.y[1, :])
pyplot.xlim(0, 5)
pyplot.ylim(0, 1)
```

# Algebra linear!!!

```python
from scipy import linalg
```

```python
A = numpy.array([[1,1], [1,2]])

b = numpy.array([2, 5])

x = linalg.solve(A, b)

print(x)
```

```python
print(numpy.dot(A, x))
```

```python
linalg.eigvals(A)
```

```python
linalg.eig(A)
```

```python
linalg.inv(A)
```

```python
linalg.det(A)
```
